#### The logistic population model with harvesting:

$x_{\mathrm{next}} = a x_{\mathrm{initial}}(1-x_{\mathrm{initial}}) - h$

where:
* $x$ is the population size between 0 and 1.
* $x_{\mathrm{initial}}$ is the current population size.
* $x_{\mathrm{next}}$ is next year's population size.
* $a$ is the ability to reproduce.
* $h$ is the size of the harvest between 0 and 1.


```python
### Code the logistic population model with harvesting

a = 2.4
h = 0.15 #helps make population stable -> subtract ie .21 every generation
x = np.zeros(50)
x[0] = 0.5
for i in range(1,50):
  x[i] = a*x[i-1]*(1-x[i-1])-h #harvest value
  x[i] = np.maximum(x[i],0) #if x[i] is negative this will replace it with 0

plt.plot(x) #code not indented so it isn't looped with above - only one final plot
plt.xlabel('Generation')
plt.ylabel('Population')
plt.ylim(0,1)
plt.xlim(0,50)
plt.show()
```

#### The logistic population model with randomness:

$x_{\mathrm{next}} = (a+r) x_{\mathrm{initial}}(1-x_{\mathrm{initial}})$

where:
* $x$ is the population size between 0 and 1.
* $x_{\mathrm{initial}}$ is the current population size.
* $x_{\mathrm{next}}$ is next year's population size.
* $a$ is the ability to reproduce.
* $r$ is a normally distributed random number that changes each year.

```python
### Code the logistic population model with randomness

a = 3.6
h = 0
x = np.zeros(50)
x[0] = 0.5
for i in range(1,50):
  r = np.random.randn()*0.7 #random number multipled by standard deviation of choice
  x[i] = (a+r)*x[i-1]*(1-x[i-1])-h #r is random namber, harvest
  x[i] = np.maximum(x[i],0) #if x[i] is negative this will replace it with 0

plt.plot(x) #code not indented so it isn't looped with above - only one final plot
plt.xlabel('Generation')
plt.ylabel('Population')
plt.ylim(0,1)
plt.xlim(0,50)
plt.show()
```

#### The logistic map:

$x_{\mathrm{next}} = a x_{\mathrm{initial}}(1-x_{\mathrm{initial}})$

where:
* $x$ is the population size between 0 and 1.
* $x_{\mathrm{initial}}$ is the current population size
* $x_{\mathrm{next}}$ is next year's population size
* $a$ is the ability to reproduce.

Follow these steps:

1. Choose a random value of $a$ from 0 to 4.

2. Choose a random starting population x_{\mathrm{initial}} from 0 to 1

3. Run the model for 2,000 generations and record the $x$ value of the last generation.

4. Repeat steps 1-3 a large number of times (thousands)

5. Make a scatter plot of $a$ (x-axis) versus final $x$ (y-axis).



```python
### Code the logistic map

a_out = [] #no defined number of array places so it can buikd with us - but if doing super large calcs it is better for code memory to define the number
x_out = []

for j in range (100000): # start with small number (ie. 10) to make sure it works then move bigger ie 1000 ->200,000 will follow structure but also have more chaos
  a = np.random.rand()*0.3+3.8 #random number between 0 and 4 / later edit was limiting numbers between 0.5 range and adding 3.3 puts them between 3.2 and 3.7
  a_out.append(a) #saves produced values to the 'a_out' array
  x = np.random.rand()  #random number between 0 and 1
  for i in range(2000): #2000 generations (loop within a loop)
    x = a*x*(1-x) #must be double indented text to be apart of the second loop (ie the 'i' loop)
  x_out.append(x) #saves produced values to the 'x_out' array

plt.scatter(a_out,x_out, s=0.7)
plt.xlabel('a')
plt.ylabel('x')
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='black')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```
