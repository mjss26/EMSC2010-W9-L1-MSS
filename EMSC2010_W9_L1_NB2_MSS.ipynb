#### The logistic population model with harvesting:

$x_{\mathrm{next}} = a x_{\mathrm{initial}}(1-x_{\mathrm{initial}}) - h$

where:
* $x$ is the population size between 0 and 1.
* $x_{\mathrm{initial}}$ is the current population size.
* $x_{\mathrm{next}}$ is next year's population size.
* $a$ is the ability to reproduce.
* $h$ is the size of the harvest between 0 and 1.


```python
### Code the logistic population model with harvesting

a = 2.4
h = 0.15 #helps make population stable -> subtract ie .21 every generation
x = np.zeros(50)
x[0] = 0.5
for i in range(1,50):
  x[i] = a*x[i-1]*(1-x[i-1])-h #harvest value
  x[i] = np.maximum(x[i],0) #if x[i] is negative this will replace it with 0

plt.plot(x) #code not indented so it isn't looped with above - only one final plot
plt.xlabel('Generation')
plt.ylabel('Population')
plt.ylim(0,1)
plt.xlim(0,50)
plt.show()
```

#### The logistic population model with randomness:

$x_{\mathrm{next}} = (a+r) x_{\mathrm{initial}}(1-x_{\mathrm{initial}})$

where:
* $x$ is the population size between 0 and 1.
* $x_{\mathrm{initial}}$ is the current population size.
* $x_{\mathrm{next}}$ is next year's population size.
* $a$ is the ability to reproduce.
* $r$ is a normally distributed random number that changes each year.

```python
### Code the logistic population model with randomness

a = 3.6
h = 0
x = np.zeros(50)
x[0] = 0.5
for i in range(1,50):
  r = np.random.randn()*0.7 #random number multipled by standard deviation of choice
  x[i] = (a+r)*x[i-1]*(1-x[i-1])-h #r is random namber, harvest
  x[i] = np.maximum(x[i],0) #if x[i] is negative this will replace it with 0

plt.plot(x) #code not indented so it isn't looped with above - only one final plot
plt.xlabel('Generation')
plt.ylabel('Population')
plt.ylim(0,1)
plt.xlim(0,50)
plt.show()
```
